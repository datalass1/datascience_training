 - server-client application; edit and run your notebooks via a web browser. The application can be executed on a PC without Internet access, or it can be installed on a remote server.
 - explanatory text or titles and subtitles to clarify code.
 - ‘magic functions’ that will make your work a lot more interactive. Check them out using %lsmagic.
 - tab completion: Simply type ```object_name.<TAB>``` to view the object’s attributes

## Keyboard Shortcuts

- ```Shift + Enter``` run the current cell, select below
- ```Enter``` take you into edit mode
- ```Esc + A``` insert cell above
- ```Esc + B``` insert cell below
- ```Esc + X``` cut selected cells
- ```Ctrl + Z``` oops I made a mistake: undo
- ```Ctrl + Y``` no mistake made: redo
- ```Esc + Y``` change the cell type to Code
- ```Esc + M``` change the cell type to Markdown

````python
%matplotlib inline 
# the output of plotting commands is displayed inline within Jupyter frontends
import matplotlib.pyplot as plt # MATLAB like
import numpy as np
````

````python
# polar plot using add_axes and polar projection
fig = plt.figure()
ax = fig.add_axes([0.0, 0.0, .6, .6], polar=True)
t = np.linspace(0, 2 * np.pi, 100)
ax.plot(t, t, color='blue', lw=3);
````
