```python
from wordcloud import WordCloud#, STOPWORDS
from PIL import Image
import numpy as np
import urllib
import requests
import matplotlib.pyplot as plt
```

```python
words = 'attention, detail, manipulate_data, success,\
communication, skills, openminded, patience, practical, inclusive,\
time, management, creativity, curiosity, pragmatic, enthusiasm,\
decisions, data, geospatial, determined, mentors, deadlines,\
networking, rolemodels, accountability, empowered, teamwork,\
empowered, patience, data, decisions, geospatial, courage, empowered,\
confidence, support, teamwork, confidence, driven, diversity,\
responsibility, calm, focus, positive, critical, analytical, diversity'
```

```python
#find a png and scrape from web for the mask (aka shape)
mask = np.array(Image.open(requests.get('http://www.clker.com/cliparts/d/5/4/a/1195423873303149451pfeil_rechts_christian_b_01.svg.med.png',
                                       stream=True).raw))
```

```python
word_cloud = WordCloud(width = 2000, height = 3000, font_path='Loma', scale=6,
                           background_color='white', mode='RGB', min_font_size=6, 
                           regexp=None, colormap='summer', mask=mask).generate(words)
plt.figure(figsize=(30, 40), dpi=300, facecolor = 'white', edgecolor='black')
plt.imshow(word_cloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
word_cloud.to_file("soft_skills1.tif")
```
